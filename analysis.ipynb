```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
%config InlineBackend.figure_format = 'svg'

import os
path = r'./data/'

colors = plt.cm.Spectral(np.arange(20))


stocks = []
close = []
pes = []
pbs = []

for file in os.listdir(path):
    if '~' in file : continue
    df = pd.read_excel(path+file)
    df.columns = ['date', 'date2', 'open', 'max', 'min', 'close', 'done', 'market_value', 'turnover', 'pe', 'pb']
    df.loc[:, 'date'] = pd.to_datetime(df.date)
    df = df.set_index('date')
    df = df.iloc[:, 1:]
   
                          
    stocks.append(file.split('.')[0])
    close.append(dict(zip(df.index.values, df.close.values)))
    pes.append(dict(zip(df.index.values, df.pe.values)))
    pbs.append(dict(zip(df.index.values, df.pb.values)))
    print(file.split('.')[0])
```

```python
colors = colors = plt.cm.Spectral(np.arange(20))
colors

plt.figure(figsize=(12,7))
for i in range(20):
    plt.plot(close[i].keys(),close[i].values(),label=stocks[i])
plt.title('start_price')
plt.legend()
plt.show()

plt.figure(figsize=(12,7))
for i in range(20):
    plt.plot(pes[i].keys(),pes[i].values(),label=stocks[i])
plt.legend()
plt.title('pe')
plt.show()

plt.figure(figsize=(12,7))
for i in range(20):
    plt.plot(pbs[i].keys(),pbs[i].values(),label=stocks[i])
plt.title('pb')
plt.legend()
plt.show()
```
